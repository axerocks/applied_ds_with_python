```python
import pandas as pd
import numpy as np

# import the csv files
"Beat","Grid","Occurence Date","Occurence Time"

colnames = ['record_id','offense_code','offense_ext','offense_category','description',
           'police_district','beat','grid','occurence_date','occurence_time']

crime_data_1year = pd.read_csv('sac_crime_data_1year.csv', 
                               skiprows  = 1,
                               names = colnames)

crime_data_2year = pd.read_csv('sac_crime_data_2year.csv', 
                               skiprows = 1,
                               names = colnames)
crime_data_2year.head(10)
```

```python
# count the number of offenses by type in each year
offenses = crime_data_1year['offense_category']
offenses_count_1year = offenses.value_counts()
offenses_count_1year = offenses_count_1year.to_frame()
offenses_count_1year.rename(columns={'offense_category': 'num_offenses_1year'}, inplace=True)

offenses = crime_data_2year['offense_category']
offenses_count_2year = offenses.value_counts()
offenses_count_2year = offenses_count_2year.to_frame()
offenses_count_2year.rename(columns={'offense_category': 'num_offenses_2year'}, inplace=True)

offenses_count_2year
```

```python
# merge the dataframes and calculate the % change
offenses_yoy = pd.merge(offenses_count_1year, offenses_count_2year, 
                        how='inner', 
                        left_index = True, 
                        right_index = True)

offenses_yoy['perc_chng'] = round((offenses_yoy['num_offenses_1year'] / offenses_yoy['num_offenses_2year']) * 100,0)

offenses_yoy
```

```python
# sort the data by the number of offenses this year and keep the top 10

offenses_yoy.sort_values(['num_offenses_1year'], ascending=[False])
top_10_offenses = offenses_yoy.head(10)

top_10_offenses.reset_index(inplace = True)
top_10_offenses.rename(columns={'index': 'offense'}, inplace=True)

top_10_offenses
```

```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
from matplotlib.figure import Figure
from textwrap import wrap


# create a new figure
fig = Figure()
plt.figure(figsize=(8,6))

plt.subplot(1, 2, 1)
plt.tight_layout()

############## create a bar chart of the top 10 offenses ################

# get data into a series
top_offenses_count = top_10_offenses['num_offenses_1year']
offense_names = top_10_offenses['offense']

index = np.arange(1,11,1)

plot = plt.bar(left = index, 
               height = top_offenses_count,
               color='lightslategrey'
               )

ax = plt.gca()

# set the x tick values
plt.xticks(np.arange(1,11), offense_names, alpha=0.8, rotation='45', size = 7)

# set the chart title
ax.set_title("\n".join(wrap('Sacramento Top 10 Crimes Last Year by Number of Offenses',40)),alpha=0.8,size = 11)


# remove the top and right border from the chart
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# direct label each bar with Y axis values

for bar in plot:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100, str('{:,.0f}'.format(height)), 
                 ha='center', fontsize=8)

plt.gcf().subplots_adjust(bottom=0.2)


############## create a line chart of percentage change ################

plt.subplot(1, 2, 2)

# remove any rows where the number of offenses last year was less than 100
offenses_yoy_gt_100 = offenses_yoy[offenses_yoy['num_offenses_1year'] > 100]

# now sort by descending percentage change
offenses_yoy_gt_100 = offenses_yoy_gt_100.sort_values(['perc_chng'], ascending=[False])

# keep all growth offenses
growth_offenses = offenses_yoy_gt_100[offenses_yoy_gt_100['perc_chng'] > 100]

growth_offenses.reset_index(inplace = True)
growth_offenses.rename(columns={'index': 'offense'}, inplace=True)
growth_offenses['perc_chng'] = growth_offenses['perc_chng'] - 100


# get data into a series
top_growth_count = growth_offenses['perc_chng']
growth_offense_names = growth_offenses['offense']

index = np.arange(1,10,1)

plot = plt.bar(left = index, 
               height = top_growth_count,
               color='lightslategrey'
               )

ax = plt.gca()

# set the x tick values
plt.xticks(np.arange(1,10), growth_offense_names, alpha=0.8, rotation='45', size = 7)

# set the chart title
ax.set_title("\n".join(wrap('Sacramento Top Growth Crimes by % Increase in Number of Offenses',40)),alpha=0.8,size = 11)


# remove the top and right border from the chart
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# direct label each bar with Y axis values
for bar in plot:
    height = bar.get_height()
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() + 3, str('{:,.0f}'.format(height)) + '%', 
                 ha='center', fontsize=8)

plt.gcf().subplots_adjust(bottom=0.2,top=0.9)

plt.savefig('sacramento_crime_stats.pdf',bbox_inches='tight',dpi=199)
```
